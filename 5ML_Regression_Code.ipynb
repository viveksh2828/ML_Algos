```python
#Decision Tree
from sklearn import tree
#from sklearn.tree import DecisionTreeClassifier 
from sklearn.tree import DecisionTreeRegressor 
#from sklearn.model_selection import train_test_split 
#from sklearn.metrics import accuracy_score 
from sklearn.metrics import mean_absolute_error
```

```python
#x=pd.DataFrame(data.iloc[:,:-1]) 
#y=pd.DataFrame(data["Outcome"]) 
x_trainDT,x_testDT,y_trainDT,y_testDT=train_test_split(x,Y,train_size=0.8,random_state=28)
print(x_trainDT.shape) 
print(x_testDT.shape) 
print(y_trainDT.shape) 
print(y_testDT.shape) 
```

```python
###### Single Decision Tree Regression in Python #######
from sklearn import tree
#choose from different tunable hyper parameters
# Splitting the dataset into training and testing set (80/20)
# Initializing the Decision Tree Regression model
model = DecisionTreeRegressor(random_state = 0)

# Fitting the Decision Tree Regression model to the data
model.fit(x_trainDT, y_trainDT)
 
```

```python
# Predicting the target values of the test set
y_predDT = model.predict(x_testDT)

# RMSE (Root Mean Square Error)
rmse = float(format(np.sqrt(mean_squared_error(y_testDT, y_predDT)), '.3f'))
print("\nRMSE: ", rmse)
```

```python
#Support Vector Regression
import pandas as pd
from sklearn import svm
RegModel = svm.SVR(C=2, kernel='linear')
 
#Printing all the parameters of KNN
print(RegModel)
 
#Creating the model on Training Data
SVM=RegModel.fit(x_trainDT,y_trainDT)
predictionSVM=SVM.predict(x_testDT)
 
#Measuring Goodness of fit in Training data
from sklearn import metrics
print('R2 Value:',metrics.r2_score(y_trainDT, SVM.predict(x_trainDT)))
 
#Measuring accuracy on Testing Data
print('Accuracy',100- (np.mean(np.abs((y_testDT - predictionSVM) / y_testDT)) * 100))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(x_trainDT, y_trainDT)
# Predicting the target values of the test set
y_predRF = regr.predict(x_testDT)
y_predRF

```

```python
# RMSE (Root Mean Square Error)
rmseRF = float(format(np.sqrt(mean_squared_error(y_testDT, y_predRF)), '.3f'))
print("\nRMSE: ", rmseRF)
```
